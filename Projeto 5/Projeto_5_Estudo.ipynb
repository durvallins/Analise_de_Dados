## Aplicando One-Hot-Encoding

https://pandas.pydata.org/docs/reference/api/pandas.get_dummies.html

```python
# Dataset original
dados
```

```python
# Obtem as variáveis dummy
df_dummies_int = pd.get_dummies(dados['Modelo']) # até aqui retornará True e False

# Convertendo True/False para 0/1
df_dummies_int = df_dummies_int.astype(int) 
print(df_dummies_int)
```

>>> A conversão da variável dummy é o pulo do gato, pois retornará 1/0 aou invés de True/False   <br>->   df_dummies_int = df_dummies_int.astype(int) 

```python
# Obtem as variáveis dummy
df_dummies_bol = pd.get_dummies(dados['Modelo'])
```

```python
# Variáveis dummy
df_dummies_bol
```

```python
df_dummies = df_dummies_int
df_dummies
```

```python
# Conectando os dfs
df = pd.concat([dados, df_dummies], axis='columns')
```

```python
# Drop da coluna catgórica
newdf = df.drop(['Modelo'], axis=1)
```

## Aplicando One-Hot-Encoding com Transformer

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html

```python
dados = pd.read_csv('dados/dataset.csv')
dados
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Cria o encoder com transformação
OHEncoder = ColumnTransformer([('Modelo', OneHotEncoder(), [0])], remainder='passthrough')
```

```python
# Cópia do df original
df_car = dados
```

```python
df_car
```

```python
# Define os valores de X
x = df_car.drop('Preco_Venda', axis = 1)
x
```

```python
y = df_car['Preco_Venda']
y
```

```python
# Aplica o encoder ao dataset x
X = OHEncoder.fit_transform(x)
X
```

```python
# Cria a versão 3 do modelo
modelo_v3 = linear_model.LinearRegression()
```

```python
# Treina o modelo
modelo_v3.fit(X, y)
```

```python
# Uma previsão do modelo
modelo_v3.predict([[0,0,1,45000,4]])
```

```python
# Uma previsão do modelo
modelo_v3.predict([[0,1,0,86000,7]])
```

```python
# Uma previsão do modelo
modelo_v3.score(X, y)
```
